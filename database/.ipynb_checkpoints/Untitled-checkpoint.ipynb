```python
csv_file_path = "Places.csv"
```

```python
import sqlite3 

with open(csv_file_path, 'r') as csv_file:
    csv_reader = csv.reader(csv_file)
    csv_file.seek(0)
    next(csv_reader)  
    for val in csv_reader:
        city = val[0]
        place = val[1].split(".")[1]
        distance = val[2]
        distancefromcitycenter = val[3]
        description = val[4]
        insert_load_world_data(city, place, distance, distancefromcitycenter, description)
```

```python
def insert_load_world_data(city, place, distance, distancefromcitycenter, description):
    connection = sqlite3.connect(database_nm)
    cur = connection.cursor()
    cur.execute("DROP TABLE IF EXISTS city_place;")
    cur.execute(sql_statement)
    cur.execute("INSERT INTO city_place (city, place, distance, distancefromcitycenter, description) VALUES (?, ?, ?, ?, ?)",
                    (city, place, distance, distancefromcitycenter, description))
    connection.commit()
    return "Record Inserted Successfully"
```

```python
sql_statement = """CREATE TABLE city_place (
    city_id INTEGER PRIMARY KEY AUTOINCREMENT,
    city TEXT NOT NULL,
    place TEXT NOT NULL,
    distance FLOAT,
    distancefromcitycenter TEXT ,
    description TEXT
);"""
```
